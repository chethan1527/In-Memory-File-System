```python
import os
import shutil

current_directory = '/'

while True:
    user_input = input(f'{current_directory} $ ')

    if user_input.startswith('mkdir'):
        _, directory_name = user_input.split()
        new_directory = os.path.join(current_directory, directory_name)
        os.makedirs(new_directory, exist_ok=True)
        print(f'Directory "{directory_name}" created.')

    elif user_input.startswith('cd'):
        _, path = user_input.split()
        if path == '..':
            current_directory = os.path.dirname(current_directory)
        elif path.startswith('/'):
            current_directory = path
        else:
            current_directory = os.path.join(current_directory, path)

    elif user_input.startswith('ls'):
        _, path = user_input.split() if len(user_input.split()) > 1 else (None, None)
        list_path = os.path.join(current_directory, path) if path else current_directory
        print('\n'.join(os.listdir(list_path)))

    elif user_input.startswith('grep'):
        try:
            _, pattern, filename = user_input.split()
            with open(os.path.join(current_directory, filename), 'r') as file:
                lines = file.readlines()
                matching_lines = [line.strip() for line in lines if pattern in line]
                print('\n'.join(matching_lines))
        except ValueError:
            print('Usage: grep pattern filename')

    elif user_input.startswith('cat'):
        _, filename = user_input.split()
        try:
            with open(os.path.join(current_directory, filename), 'r') as file:
                content = file.read()
                if content:
                    print(content)
                else:
                    print(f'File "{filename}" is empty.')
        except FileNotFoundError:
            print(f'Error: File "{filename}" not found.')

    elif user_input.startswith('touch'):
        _, filename = user_input.split()
        with open(os.path.join(current_directory, filename), 'w'):
            pass
        print(f'File "{filename}" created.')

    elif user_input.startswith('echo'):
        _, rest = user_input.split(' ', 1)
        text, filename = rest.split(' > ')
        with open(os.path.join(current_directory, filename), 'w') as file:
            file.write(text)
        print(f'Text written to "{filename}".')


    elif user_input.startswith('mv'):
        _, source, destination = user_input.split()
        shutil.move(os.path.join(current_directory, source), os.path.join(current_directory, destination))
        print(f'Moved "{source}" to "{destination}".')

    elif user_input.startswith('cp'):
        _, source, destination = user_input.split()
        shutil.copy2(os.path.join(current_directory, source), os.path.join(current_directory, destination))
        print(f'Copied "{source}" to "{destination}".')

    elif user_input.startswith('rm'):
        _, target = user_input.split()
        try:
            target_path = os.path.join(current_directory, target)
            if os.path.isfile(target_path):
                os.remove(target_path)
            elif os.path.isdir(target_path):
                shutil.rmtree(target_path)
            print(f'Removed "{target}".')
        except FileNotFoundError:
            print(f'Error: "{target}" not found.')
    elif user_input.startswith('exit'):
        print('Exiting the program.')
        break

    else:
        print('Invalid command. Supported commands: mkdir, cd, ls, grep, cat, touch, echo, mv, cp, rm,exit')
```
